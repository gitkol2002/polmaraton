Ta cześć zadania wysyła wszystki pliki z katalogu upload do Cloud Storage


```python
from dotenv import load_dotenv

load_dotenv()
```

```python
import boto3
import os

s3 = boto3.client(
    "s3",
   )
```

```python
BUCKET_NAME = "maraton"
```

```python
upload_dir = "upload"

for root, dirs, files in os.walk(upload_dir):
    for file in files:
        local_path = os.path.join(root, file)

        # nazwa pliku w S3 = TYLKO nazwa pliku (bez ścieżki)
        s3_key = os.path.basename(file)

        print(f"Wysyłam plik: {local_path} ➜ s3://{BUCKET_NAME}/{s3_key}")

        s3.upload_file(
            Filename=local_path,
            Bucket=BUCKET_NAME,
            Key=s3_key
        )

print("✓ Wszystkie pliki zostały wysłane poprawnie!")
```
